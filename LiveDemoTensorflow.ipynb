```python
import cv2
import numpy as np
import tensorflow as tf
import cv2
import os 
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "haarcascade_frontalface_default.xml"
# I can change the font scale and font to what I want
font_scale = 1.5
font = cv2.FONT_HERSHEY_PLAIN
new_model = tf.keras.models.load_model("my_model1.h5")
```

```python
# set the rectangle background to white
rectangle_bgr =(255,255,255)
```

```python
# make a black image
# creating a canvas of 500x500 and setting it all to black
img = np.zeros((500,500))
```

```python
# Set some text
text = "Some text in box!"
# get the width and height of the text box
(text_width,text_height) = cv2.getTextSize(text, font, fontScale = font_scale, thickness = 1)[0]
# Set text start position of tect that willmbe overlad onto our live image
# img.shape[0] returns the height of the image, which is used to calculate the y coordinate of the starting position for the text. 
# By subtracting 25 from the height, the text is positioned close to the bottom of the image, leaving some space for other annotations
# or visual elements.
text_offset_x = 10
text_offset_y = img.shape[0] - 25

# Make the co ordinates of the box with a small padding of two pixels
# cv2.rectangle() is an OpenCV function that draws a rectangle on an image. Here, the function is called with the following arguments:
# img: the image to draw on
# box_coords[0]: the top-left corner of the rectangle
# box_coords[1]: the bottom-right corner of the rectangle
# rectangle_bgr: the color of the rectangle
# cv2.FILLED: the thickness parameter, which specifies that the rectangle should be filled in (rather than just an outline)
box_coords = ((text_offset_x,text_offset_y),(text_offset_x + text_width +2, text_offset_y - text_height -2))
cv2.rectangle(img, box_coords[0], box_coords[1], rectangle_bgr,cv2.FILLED)

# cv2.putText() is an OpenCV function that adds text to an image. Here, the function is called with the following arguments:
# img: the image to add text to
# text: the text to add
# (text_offset_x,text_offset_y): the starting position of the text
# font: the font type to use for the text
# fontScale: the font scale factor, which determines the size of the text
# color: the color of the text
# thickness: the thickness of the text stroke (i.e., the outline)
cv2.putText(img, text, (text_offset_x,text_offset_y), font, fontScale=font_scale, color=(0,0,0), thickness=1)
```

```python


cap = cv2.VideoCapture(0)
# Check if webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")
```

```python

    
while True:
    ret,frame = cap.read()
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray,1.1,4)
#     Checking all possible faces in the current frame
    for x,y,w,h in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2)
        facess = faceCascade.detectMultiScale(roi_gray)
        if len(facess)==0:
            print("Face not detected")
        else:
            for(ex,ey,ew,eh) in facess:
                face_roi = roi_color[ey:ey+eh, ex:ex +ew] ##Cropping the face
    
    final_image = cv2.resize(face_roi, (224,224))
    The reason for adding an extra dimension to the image in this way is often because many deep learning models, such as convolutional neural networks (CNNs), expect their inputs to be four-dimensional tensors, with the first dimension representing the batch size. In this case, adding the extra dimension allows the image to be used as input to a CNN.
    final_image = np.expand_dims(final_image, axis = 0) #need fourth dimension
    final_image = final_image/255.0
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    Predictions = new_model.predict(final_image)
    
    font_scale = 1.5
    font = cv2.FONT_HERSHEY_PLAIN
    
    
    if (np.argmax(Predictions)==0):
        status = "Angry"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
#         Writes the 
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    elif (np.argmax(Predictions)==1):
        status = "Disgust"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    elif (np.argmax(Predictions)==2):
        status = "Fear"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    elif (np.argmax(Predictions)==3):
        status = "Happy"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))

    elif (np.argmax(Predictions)==4):
        status = "Sad"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    elif (np.argmax(Predictions)==5):
        status = "Surprise"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    else:
        status = "Neutral"
        
        x1,y1,w1,h1 = 0,0,175,75
        #Draw black background rectangle
        cv2.rectangle(frame, (x1,x1), (x1 +w1, y1 +h1), (0,0,0), -1)
        #Add text
        cv2.putText(frame, status, (x1 +int(w1/10), y1 + int(h1/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
        
        cv2.putText(frame, status, (100,150), font, 3,(0,0,255),2,cv2.LINE_4)
        
        cv2.rectangle(frame,(x,y), (x+w, y+h), (0,0,255))
        
    
    cv2.imshow("Face EMotion Recognition", frame)
    
    if cv2.waitKey(2) & 0xFF ==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
    
```
